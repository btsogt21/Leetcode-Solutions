## Description

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.
<pre>
Symbol       Value <br/> 
I             1<br/> 
V             5<br/> 
X             10<br/> 
L             50<br/> 
C             100<br/> 
D             500<br/> 
M             1000
</pre>
For example, 2 is written as II in Roman numeral, just two one's added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 

X can be placed before L (50) and C (100) to make 40 and 90. 

C can be placed before D (500) and M (1000) to make 400 and 900.

Given an integer, convert it to a roman numeral.


## Example 1:

Input: num = 3

Output: "III"

Explanation: 3 is represented as 3 ones.

## Example 2:

Input: num = 58

Output: "LVIII"

Explanation: L = 50, V = 5, III = 3.

## Example 3:

Input: num = 1994

Output: "MCMXCIV"

Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.

## Constraints:

1 <= num <= 3999

```python
from __future__ import annotations #this was imported so that I could use built in types as generics. 
# Only >3.9 versions of python can use built in types as generics without this import.
```

```python
# First accepted solution. This one was written without assistance.

# This problem was pretty simple too. We essentially have an alternative number system that isn't too far off from arabic numerals,
# and we need to convert input ints into roman numeral strings. To do this, I first created a dictionary containing, in descending order,
# the full set of roman numerals we are expected to use in this problem. Then, I create a loop to divide the input int by each 
# of these roman numerals' values one by one, making sure to append to a return string the corresponding letter of each roman 
# numeral as many times as said numeral fits inside the given int. Moreover, if the numeral fits inside the input int at all, we 
# make sure to modulus the input int by the roman numeral so as to remove the number on the leftmost of the input int (also akin
# to simply subtracting the (number of times the current roman numeral fits inside current input int)*(value of roman numeral) 
# from the input int). After letting this loop run through the entire dictionary, we return the return string.


class Solution:
    def intToRoman(self, num: int) -> str:
        return_str = ''
        numeral_dict={
            'M':1000,
            'CM':900,
            'D':500,
            'CD':400,
            'C':100,
            'XC':90,
            'L':50,
            'XL':40,
            'X':10,
            'IX':9,
            'V':5,
            'IV':4,
            'I':1
        }
        for i in numeral_dict:
            save = num//numeral_dict[i]
            if save>0:
                num%=numeral_dict[i]
                #num-=save*numeral_dict[i] another option
            return_str+=(i*save)
        return(return_str)
    
    
# Time complexity should be constrained by the number of decimal places inside the input string and by the size of our numeral
# dictionary, but since we only allow input sizes up to 3999 (four decimal places), and since our numeral dictionary has a set 
# size, time complexity should essentially be O(1).
```

```python
# Second accepted solution. This one was written to be faster by utilizing two arrays for our key value pairs instead of a 
# dictionary. 

class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        
        romans_digits = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        romans_letters = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
       
        roman = ""
        for i in range(len(romans_digits)):
            while num >= romans_digits[i]:
                roman += romans_letters[i]
                num -= romans_digits[i]
        return roman
# Time complexity is again limited by the size of the arrays and the number of decimal places in the input number.
```
